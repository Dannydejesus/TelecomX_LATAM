#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
import json
```

```python
df = pd.read_json('/content/TelecomX_Data.json')
df.head()
```

```python
# Lista de columnas con datos anidados
columnas_a_normalizar = ['customer', 'phone', 'internet', 'account']

# Normalizar cada columna sin perder registros
for columna in columnas_a_normalizar:
    if columna in df.columns:
        datos_normalizados = pd.json_normalize(df[columna])
        df = df.drop(columns=[columna]).reset_index(drop=True).join(datos_normalizados)
```

```python
data_normalizada = df
data_normalizada.head()
```

#ðŸ”§ TransformaciÃ³n

```python
data_normalizada.sample(6)
```

```python
data_normalizada['Churn'] = data_normalizada['Churn'].replace('', 'no')
data_normalizada['Churn'] = data_normalizada['Churn'].replace('no', 'No')

columnas_binarias = ["Churn", "Partner", "Dependents", "PhoneService", "PaperlessBilling"]
data_normalizada[columnas_binarias] = data_normalizada[columnas_binarias].replace({"Yes": 1, "No": 0}).astype('int8')

columnas_categoricas = ['gender', 'InternetService', 'Contract', 'PaymentMethod']
data_normalizada[columnas_categoricas] = data_normalizada[columnas_categoricas].astype('category')

columnas_servicios = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                'TechSupport', 'StreamingTV', 'StreamingMovies']
data_normalizada[columnas_servicios] = data_normalizada[columnas_servicios].replace({
    'No internet service': 'No',
    'No': 0,
    'Yes': 1
}).astype('int8', errors='ignore')

data_normalizada['MultipleLines'] = data_normalizada['MultipleLines'].replace({
    'No phone service': 'No',
    'No': 0,
    'Yes': 1
}).astype('int8', errors='ignore')

data_normalizada['SeniorCitizen'] = data_normalizada['SeniorCitizen'].astype('int8')
data_normalizada['Charges.Total'] = pd.to_numeric(data_normalizada['Charges.Total'], errors='coerce')
data_normalizada['Charges.Total'] = data_normalizada['Charges.Total'].fillna(data_normalizada['tenure'] * data_normalizada['Charges.Monthly'])
# crear la columna Cuentas_Diarias.
data_normalizada['Cuentas_Diarias'] = (data_normalizada['Charges.Monthly'] * 12 / 365).round(2)
```

```python
# Diccionario de traducciÃ³n de nombres de columnas
traducciones = {
    'customerID': 'ID_Cliente',
    'Churn': 'Evasion',
    'gender': 'Genero',
    'SeniorCitizen': 'Ciudadano_Mayor',
    'Partner': 'Pareja',
    'Dependents': 'Dependientes',
    'tenure': 'Antiguedad_Meses',
    'PhoneService': 'Servicio_Telefonico',
    'MultipleLines': 'Lineas_Multiples',
    'InternetService': 'Servicio_Internet',
    'OnlineSecurity': 'Seguridad_En_Linea',
    'OnlineBackup': 'Respaldo_En_Linea',
    'DeviceProtection': 'Proteccion_Dispositivo',
    'TechSupport': 'Soporte_Tecnico',
    'StreamingTV': 'TV_Streaming',
    'StreamingMovies': 'Peliculas_Streaming',
    'Contract': 'Tipo_Contrato',
    'PaperlessBilling': 'Facturacion_Digital',
    'PaymentMethod': 'Metodo_Pago',
    'Charges.Monthly': 'Cargos_Mensuales',
    'Charges.Total': 'Cargos_Totales',
    'Cuentas_Diarias': 'Cuentas_Diarias'
}

# Aplicar las traducciones
data_normalizada = data_normalizada.rename(columns=traducciones)

# Verificar resultados
print("Columnas traducidas:")
print(data_normalizada.columns.tolist())
```
